```python
#@title Default title text
# ml modeling and evaultion 
dot1 = graphviz.Digraph( comment='ML classification model') 
dot1.node('A', 'Start',shape="oval") 
dot1.node('B', 'Building the dataset',shape="box3d")
dot1.node('C', 'Data annotation',shape="diamond")
dot1.node('D', 'Data sources (Twitter Data)',shape="cylinder")
dot1.node('E', "{Data Preprocessing|{Tokenization|Vectorizer|Lemmatization}}",shape="Mrecord")
dot1.node('X', "Pre-processed data",shape="parallelogram")
dot1.node('F', 'Data Visualization',shape="box3d")
dot1.node('G', "Word Cloud|Matplotlib|Seaborn",shape="Mrecord")
dot1.node('H', 'Feature Extraction',shape="box3d")
dot1.node('I', 'TF-IDF Weighting',shape="rect")
dot1.node('J', 'Embedding Matrix',shape="parallelogram")
dot1.node('Y', 'Classification & Evaluation',shape="box3d")
dot1.node('K', 'Splitting Data',shape="rect")  
dot1.node('L', 'Training data',shape="parallelogram")  
dot1.node('M', 'Testing data',shape="parallelogram") 
dot1.node('N', 'Training using ML Classification Model',shape="parallelogram")  
dot1.node('O', '{Random Forest |logistic Regression|SVM|RNN}',shape="Mrecord")  
dot1.node('P', 'Testing Model',shape="rect")  
dot1.node('Q', 'Saving Model',shape="rect")  
dot1.node('R', 'End',shape="oval")  

dot1.edges(['AB','BC','CE','EX','XF','FG','HJ','JN','XY','XH','KL','KM','LN','MN','ON','NP','PQ','QR','YK'])
dot1.edge('D','C',constraint="false")
dot1.edge('J','I',dir="back")


dot1.render('test-output/dot1.jpg', view = False)
dot1
```

```python
#@title Default title text
dot = graphviz.Digraph( comment='Hate_speech chatbot')  
dot.node('A', 'Start',shape="oval")  
dot.node('B', 'Input data',shape="rect")
dot.node('C', 'List of services',shape="diamond")
dot.node('D', 'Enter Text',shape="parallelogram")
dot.node('E', 'Upload image',shape="parallelogram")
dot.node('F', 'Extract text',shape="square")
dot.node('G', 'Data Processing',shape="rect")
dot.node('H', 'Applying classification model on the user data',shape="rect")
dot.node('I', 'Result',shape="rect")



dot.node('K', 'Is hate speech/ offensive / none ?',shape="diamond")
dot.node('O', 'Save into database',shape="oval")
dot.node('L', 'Displays a warning: This is hate speech',shape="parallelogram")
dot.node('M', 'Displays a warning: This is offensive speech',shape="parallelogram")
dot.node('Q', 'Displays : This is a safe speech',shape="parallelogram")
dot.node('N', 'Continue ?',shape="diamond")


dot.node('P', 'End',shape="oval")  
dot.edges(['AB', 'BC','DG','CE','EF','FG','GH','HI','IK','LN','MN','QN'])
dot.edge('C','D',constraint='false')

dot.edge('I','O', constraint ='true')
dot.edge('K', 'L', 'Hate')
dot.edge('K', 'M', 'Offensive')
dot.edge('K', 'Q', 'None')
dot.edge('N', 'B', 'YES',constraint='false')
dot.edge('N', 'P', 'NO')
dot.render('test-output/dot.jpg', view = False)
dot
```
